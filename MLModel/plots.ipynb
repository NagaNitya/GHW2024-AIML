Matplotlib plots

Importing package and setting up data

```python
%matplotlib inline
from matplotlib import pyplot as plt

x=[5, 2, 9, 4, 7, 6, 1, 8] # x values
y=[10, 5, 8, 4, 2, 3, 7, 9] # y values
```

line plot

```python
plt.plot(x, y, linestyle='--', color='r', marker='o', label='Line 1')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

Bar Plot

```python
plt.bar(x, y)
plt.show()
```

Horizontal Bar Plot

```python
plt.barh(x, y)
plt.show()
```

Histogram

```python
plt.hist(x)
plt.show()
```

Scatter Plot

```python
plt.scatter(x, y)
plt.show()
```

Pie Graph

```python
# data
labels=['s1', 's2', 's3', 's4', 's5', 's6', 's7', 's8']
sizes=[20, 15, 10, 25, 20, 5, 5, 10]

# chart
plt.title('Pie Chart')
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, explode=[0.2, 0.1, 0.5, 0.2, 0.3, 0.4, 0.6, 0.1], shadow=True) # pie chart
plt.show()
```

Matrix PLot

```python
# data
datamat=[[5, 2, 9, 4, 7, 6, 1, 8], [10, 5, 8, 4, 2, 3, 7, 9], [20, 15, 10, 25, 20, 5, 5, 10]]

# chart
plt.matshow(datamat) # matrix plot
plt.show()

# darkest=0
# lightest=largest number
```
